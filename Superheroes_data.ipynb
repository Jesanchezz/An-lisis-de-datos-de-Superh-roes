Análisis de datos de tipo descriptivo para decribir y presentar de manera detallada datos relevantes y significativos sobre personajes icónicos de Marvel, DC y George Lucas

```python
__Author__ = 'JesusD'
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Superheroes.csv')

print(df)
```

Se carga el archivo .CVS al archivo .ipynb y se transforma a un DataFrame. Se visulizan los primeros registros para tener una idea rápida sobre a que nos enfretamos. 

```python
df.info()
```

Solicitamos información para tener detalles más específicos del conjunto de datos

```python
df.describe()
```

Obtenemos una breve descripción y comprobamos a cabalidad cuantos valores nulos existen en nuestro conjunto de datos. Que desde un princiío ya se había visualizado.

```python
df.isnull()
```

Se Inicia el Poceso de Limpieza de datos

```python
df_cleaned = df.dropna()
df_cleaned = df_cleaned.drop_duplicates()

columns = ['Intelligence','Name','Omnipotent','Omnipresent','Omniscient',
           'Collections','Class_value','Speed_velocity','Strength_force',
           'History', 'Tier']

df_cleaned = df_cleaned.drop(columns, axis=1)
df_cleaned.info()
```

Se eliminan los valores duplicados, nulos y eliminamos una series de columnas que no aportan información para este estudio. No se elimnan todas porque se guardará el Dataframe limpio en un archivo .CSV para otros estudios.

```python
print(df_cleaned['Character'])
```

Se visualiza, los datos de nuestra columna ['Character'], que será una columna protagosnista en nuestro caso, por lo que hay que ver si existen valores basura, todavia. Es evidente, que todavía hay valores duplicados, que se eliminarán en un instante.

```python
df_cleaned = df_cleaned.drop_duplicates(subset = ['Character'], keep='last')
df_cleaned.info()
```

Ahora, volvemos a verificar alguna anomalia. No obstante, se aplica a la columna ['Power'] un diagrama de dispersión para visualizar su distribución y detectar alguna anomalía.

```python
plt.figure(figsize=(7,5))
plt.scatter(range(len(df_cleaned['Character'])),df_cleaned['Power'], 
            label='Datos de dispersión', color ='b', marker='o')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Diagrama de dispersión')
plt.show()
```

Las distribuciones que se visualizan en el gráfico, revelan que no hay ninguna anomalía ya que se agrupan en rangos iguales y uno que otro que se encuentra solitario, pero es es un rango aceptable. 

Análisis, Representación Gráfica y Resultados

```python
#Número de Héroes de marvel y DC (Gráfico de Columnas)
count = df_cleaned['Creator'].value_counts()
plt.figure(figsize=(7, 5))
bars = plt.bar(count.index, count.values, color=['blue', 'red', 'navy'])
plt.xlabel('Creator')
plt.ylabel('Counts')
plt.title('Creator Heroes')
plt.xticks(rotation=45) 
plt.bar_label(bars, fmt='%d', label_type='edge')
plt.show()
```

*Pregunta: ¿Cuántos héroes de Marvel, DC y George Lucas hay en el conjunto de datos?

En este gráfico se muestra el número de héroes del conjunto de datos agrupados según su creador. Hay 3 creadores: Marvel, DC y George Lucas.

*Conclusiones:

    -Marvel es creador con mayor número de héroes en el conjunto de datos, con 16 héroes.
    -En segundo Lugar esta DC, con 10 héroes en el conjunto.
    -George Lucas, es el creador con menos héroes en el conjunto de datos (de este estudio), solo tiene un solo héroe.

```python
#Los 5 héroes más poderosos (gráfico de Columnas)

df_cleaned['Power'] = df_cleaned[['Power']].apply(lambda x: x.astype(int))
powerfull_heroes = df_cleaned[df_cleaned['Power'] >= 80]
powerfull_heroes = powerfull_heroes.sort_values('Power')
powerfull_heroes = powerfull_heroes.head(5)

plt.figure(figsize= (8,5))
bars = plt.bar(powerfull_heroes['Character'], powerfull_heroes['Power'] , color = 'Orange')
plt.title('The 5 Most Powerfull Heroes')
plt.xlabel('Character')
plt.ylabel('Power')
plt.bar_label(bars, label_type= 'edge')
plt.xticks(rotation=40) 
plt.show()
```

*Pregunta: ¿Cuáles son los 5 héroes más poderosos del conjunto de datos?

Para identificar estos 5 héroes, se filtró de acuerdo a su poder, partiendo de la premisa que un héroe poderoso es aquel cuyo poder es mayor o igual a un nivel de poder de 80. A partir de esta condición se filtró a esos héroes y se limitó el resultado a solo 5 que correspondiera a esta condición, ordenandolos en orden ascendente en el gráfico de barra.

*Conclusión:

    -Los héroes más poderosos de los 5 son Thor y Flash tienen un poder de 100.
    -Le sigue Capitana Marvel con un poder de 90. 
    -Y finalmente, Iron man y Lex Luthor  con un poder de 80.
    -Hay una exclusión de los creadores en este caso, cosa que se podría modificar si se quisiese.

```python
#Número de Héroes con ojos azules, verde y marrones
eye_color_heroes = df_cleaned[df_cleaned['Eye_color'].isin(['Green', 'Brown', 'Blue'])]
counts = eye_color_heroes['Eye_color'].value_counts()
plt.figure(figsize=(7,4))
barh = plt.barh(counts.index, counts.values, color=['blue','brown','green'])
plt.bar_label(barh, labels=counts.values)
plt.xlabel('Ocurrence number')
plt.ylabel('Eye Color')
plt.title('Heroes eye color')
plt.show()
```

*Pregunta: ¿Cuántos Héroes tienen los ojos azules, verdes y marrones correspondientemente?

Se realizó un gráfico de barras horizontales para representar visualmente los resultados. Se filtró del conjunto de datos aquellos datos que correspondienran y se le aplico un contador para tener un conteo de cada ocurrencia que aplicar a una de estas tres condiciones.

*Conclusiones:

    -Hay más presencia de héroes con los ojos azules ya que hay 16.
    -Hay muy pocos héroes con los verdes, solo hay 3.
    -La suma total de los héroes con estos colores de ojos, da 25... ¿Cuál sería el otro color de ojos que tienen los heroés de nuestro conjunto de datos?

```python
#Porcentaje de héroes másculinos y femeninos (gráfico circular)
total = df_cleaned['Gender'].value_counts().sum()
male = df_cleaned[df_cleaned['Gender'].isin(['Male'])].value_counts().sum()
percentage_male = (male * 100) / total
female = df_cleaned[df_cleaned['Gender'].isin(['Female'])].value_counts().sum()
percentage_female = (female * 100) / total

plt.pie([percentage_male, percentage_female], labels= ['Male', 'Female'], 
        autopct='%1.1f%%', startangle=140, colors=['orange', 'skyblue'])  
plt.axis('equal') 
plt.title('Percentage Heroes Male/Female')
plt.legend()
plt.show() 
```

*Pregunta: ¿Cuál es el porcentaje héroes masculinos y femeninos del conjunto de datos?

Para determinar ese porcentaje se procedió primero a identificar el porcentaje de héroes masculino y femenino. Una vez que se obtuvo esta información se dispuso en un gráfico circular dichos porcentajes.

*Conclusión: 

    - Hay mayor porcentaje de héroes masculinos en el conjunto de datos (81,5 %).
    - La suma del porcentaje de héroes masculinos (81,5 %) más el porcentaje de héroes femeninas (18,5 %) da como resultado 100%, lo que significa que el cálculo porcentual con su distribución son correctos.
